# FEATURE SCALING

```python
from sklearn import *
max_scaling=preprocessing.MinMaxScaler()
x=max_scaling.fit_transform(x1)
```

```python
x=pd.DataFrame(x,columns=x1.columns)
```

```python
x
```

# for balanced data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
nm=RandomOverSampler(random_state=100)
```

```python
X,Y=nm.fit_resample(x,y)
```

```python
Y.value_counts()
```

# Data UnderFit or OverFit 

```python
from xgboost import XGBClassifier
```

```python
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(X,Y,test_size=.25)
```

```python
train_score=[]
test_score=[]
depth=[n for n in range(1,16)]
```

```python
for i in depth:
    #dtc=tree.DecisionTreeClassifier(max_depth=i)
    dtc=XGBClassifier(max_depth=i)
    dtc.fit(xtrain,ytrain)
    test_acc=dtc.score(xtest,ytest)
    train_acc=dtc.score(xtrain,ytrain)
    train_score.append(train_acc)
    test_score.append(test_acc)
    print('Depth = %d , Test = %f , Train = %f ' % (i,test_acc,train_acc) )
```

```python
print('Max score: ','Test = %f , Train = %f ' % (test_acc.max(),train_acc.max()))
```

```python
import matplotlib.pyplot as plt
plt.plot(depth,train_score, '-o',label='Train')
plt.plot(depth,test_score, '-o',label='Test')
plt.xlabel('Depth')
plt.ylabel('Train/Test Accuracy')
plt.legend()
```
